```python
# Store columns with less than 50% missing values
dataset_rmcols= dataset[dataset.columns[dataset.isnull().sum()/dataset.shape[0]<0.5]]
```

```python
dataset_rmcols.head()
```

```python
dataset.describe(include = 'all').style
```

```python
def color_negative_red(val):
    color = 'red' if val < 0 else 'black'
    return 'color: %s' % color
```

```python
s=dataset_rmcols.describe().style.applymap(color_negative_red)
s
```

```python
#Formatting statistics table for numeric variables 
cm = sns.light_palette("green", as_cmap=True)
s.background_gradient(cmap=cm)
```

```python
#Formatting statistics table for categorical variables 
dataset_rmcols.describe(include=['O']).style.set_properties(**{'background-color': 'grey',
                           'color': 'white',
                           'border-color': 'white'})
```
